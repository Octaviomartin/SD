```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data_train, data_test, output_train, output_test = train_test_split(
    data, output, test_size=0.2, random_state=0)

print("Train sample shape :", data_train.shape)
print("Test sample shape :", data_test.shape)

```

## Q4:

```python
cov_x = np.cov(np.transpose(data_train))
eigen_values, eigen_vectors = np.linalg.eigh(cov_x)

sort_values = np.array(list(reversed(eigen_values)))
sort_vectors = np.array(list(reversed(eigen_vectors)))

n = len(sort_values)

X = np.linspace(1,60,num=60)

plt.plot(X,sort_values[0:60],"o")
plt.ylabel("eigenvalues")
plt.show()
```

We only keep the significant eigenvalues. Moreover there won't be more than 60 significant eigenvalues. The feature vector will therefore only contain the eigen vector corresponding with the highest eigenvalues.

## Q5:

## Q8:

```python

data_train_augmented = np.column_stack((np.ones((data_train.shape[0], 1 )), data_train))
data_train_augmented



p = data_train_augmented.shape[1]
n = data_train_augmented.shape[0]

gram = np.dot(data_train_augmented.T,data_train_augmented)/n

variables_left = list(range(p))
variables_added = []

#The residual
r = output_train 

#Matrix of the t-tests results
tests = np.zeros((p,p))



#The intercept is already a variable
model = LinearRegression(fit_intercept = False)


for k in range(p):
    #matrix of the potential residuals
    next_r = np.zeros((n,p))
    for var in variables_left:
        ek = np.zeros(p)
        ek[k]=1
        #Linear regression model for each covariates
        model.fit(data_train_augmented[:,[var]], r)
        #Compute the potential residual
        next_r[:,var] = r - model.predict(data_train_augmented[:,[var]])
        
        sigma = np.sum(next_r[:,var]**2)/(n-2)
        snk_2 = n/np.sum(data_train_augmented[:,var]**2)
        
        
        # We compute the value of the t-test
        tests[k, var] = np.sqrt(n) * np.abs(model.coef_)/np.sqrt(sigma*snk_2)
        
    
    best_v = np.argmax(tests[k,:])
    
    variables_left.remove(best_v)
    variables_added.append(best_v)
    r = next_r[:,best_v]
    
```

## Q9:

```python
#Plot the three graphs of the test statistics obtained when selecting the 1st, 2nd and 3rd variables
for k in range(3):
    plt.plot(np.arange(p),tests[k,:], label = "step " + str(k+1))
    plt.axis(xmin = -1, xmax = 210, ymin = -1)
plt.legend()
plt.title("Values of the t-stat at steps 1,2,3")
plt.xlabel("Covariates")
plt.show()
```

```python
from scipy.stats import norm
p_vals = []
i=-1
selected_variables = []
for var in variables_added:
    i+=1
    
    p_vals.append(2*(1-norm.cdf(tests[i,var])))
    if p_vals[i]<.1:
        selected_variables.append(var)
        
selected_variables=np.array(selected_variables)

plt.plot(np.arange(p), p_vals, 'o')
plt.plot([-0.5,50], [.1,.1], color = "black")
plt.axis(xmin = -1, xmax = 50, ymin = -0.1)
plt.title("plot of the p-values")
plt.xlabel("steps")
plt.ylabel("p-values")
plt.show()
print(selected_variables)
```

```python
#Generate the folds
n = data_train.shape[0]


index = np.array(data_train.index)

np.random.shuffle(index)

rd_index = []


   
rd_index.append(index[0:88])

rd_index.append(index[88:176])

rd_index.append(index[176:264])

rd_index.append(index[264:n])
 
    
X = []
Y = []


for k in range(4):
    X.append(data_train.loc[rd_index[k]])
    Y.append(output_train.loc[rd_index[k]])
    print(k,X[k].shape[0], "lines")
    
```

```python
from sklearn.model_selection import KFold
```

```python
#Generating the folds using KFOLDS

kf = KFold(n_splits=4)
sum = 0
X_train = []
X_test = []
Y_train = []
Y_test = []
for i,(train, test) in enumerate(kf.split(data_train)):
    X_train.append(data_train.loc[train])
    X_test.append(data_train.loc[test])
    Y_train.append(output_train.loc[train])
    Y_test.append(output_train.loc[test])
    X_train[i] = X_train[i].dropna()
    X_test[i] = X_test[i].dropna()
    Y_train[i] = Y_train[i].dropna()
    Y_test[i] = Y_test[i].dropna()
```

## Q12:

```python
from sklearn.linear_model import Ridge
```

```python

#I used my second way of splitting despite the fact that I had to get rid of NaN values, 
#and thus reduced the number of rows
alphas = [i * 10**j for j in range(-9,3) for i in range(1,10)]
risk = np.zeros(len(alphas))
score = np.zeros(len(alphas))


for k in range(4):    
    for i in range(len(alphas)):
        ridge_k = Ridge(alphas[i]).fit(X_train[k], Y_train[k])
        risk[i] += np.sum((ridge_k.predict(X_test[k]) - Y_test[k])**2)/len(Y_test[k])
        score[i] += ridge_k.score(X_test[k],Y_test[k])
        
    
```

```python
plt.plot(alphas, risk/4)
plt.xscale("log")
plt.ylabel("Risk")
plt.xlabel("Lambda")
plt.show()

plt.plot(alphas, score/4)
plt.xscale("log")
plt.ylabel("R2")
plt.xlabel("Lambda")
plt.show()
```

```python
risk_optimal = min(risk)/4
alpha = alphas[np.argmin(risk)]
score_alpha = score[np.argmin(risk)]/4
score_max = max(score)/4



print("Chosen parameter :", alpha)
print("Risk associated :", risk_optimal)
print("R2 associated :", score_alpha)
print()
print("Maximum R2 observed :", score_max)
```

```python
model = Ridge(alpha).fit(data_train,output_train)
output_predict = model.predict(data_test)

print('----Results----\n')
print('Determination Coefficient:',model.score(data_test,output_test),'\n')
print('Risk :', np.sum((output_predict - output_test)**2)/len(output_test))
```
